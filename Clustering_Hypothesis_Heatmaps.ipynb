```python
path = '/home/kaled/NeuroMat/Clustering/NeuroMat_Clustering_EEG'
data_dir = pjoin(dirname(sio.__file__), 'matlab', path)
```

```python
mat_fnames_origin_qua = [pjoin(data_dir, 
                               'CompDendro/Origin_Qua/'+filename+'.mat') for filename in filenames_origin]
mat_fnames_origin_ter = [pjoin(data_dir, 
                               'CompDendro/Origin_Ter/'+filename+'.mat') for filename in filenames_origin]
mat_fnames_pitch_qua = [pjoin(data_dir, 
                              'CompDendro/Pitch_Qua/'+filename+'.mat') for filename in filenames_pitch]
mat_fnames_pitch_ter = [pjoin(data_dir, 
                              'CompDendro/Pitch_Ter/'+filename+'.mat') for filename in filenames_pitch]
```

```python
mat_contents_origin_qua = [sio.loadmat(mat_fname) for mat_fname in mat_fnames_origin_qua]
mat_contents_origin_ter = [sio.loadmat(mat_fname) for mat_fname in mat_fnames_origin_ter]
mat_contents_pitch_qua = [sio.loadmat(mat_fname) for mat_fname in mat_fnames_pitch_qua]
mat_contents_pitch_ter = [sio.loadmat(mat_fname) for mat_fname in mat_fnames_pitch_ter]
```

```python
def heatmap_print(idx, filenames_origin, mat_content):
    print(filenames_origin[idx])
    
    dis_mat = mat_content['dis_mat']
    
    for i in range(len(dis_mat)):
        dis_mat[i][i] = 1
        
    link__ = linkage(dis_mat, method='complete')
    aris_DF = pd.DataFrame(mat_content['dis_mat'])
    sns.clustermap(aris_DF, metric="euclidean", method="complete", row_linkage=link__, col_linkage=link__,
                   cmap="PRGn", col_cluster=True)
    plt.show()
```

## Origin - Quatern√°ria

```python
for idx, mat_content in enumerate(mat_contents_origin_qua):
    
    heatmap_print(idx, filenames_origin, mat_content)
```

## Orgin - Tern√°ria

```python
for idx, mat_content in enumerate(mat_contents_origin_ter):
    heatmap_print(idx, filenames_origin, mat_content)
```

## Pitch - Quatern√°ria
